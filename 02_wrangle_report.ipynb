# Data Assessing

This stage is one of the key to achieving our end goal. After viewing the entire data using text editors. We programmatically looked through the data and we identified some quality and tidiness issues which needed to be addressed.

#Data Cleaning

One of the first thing we did was to remove tweets that only constituted replies/comments/retweets as they do not form the basis of our analysis i.e rating dog images. As a result, we were left with missing values in the columns concerned so we dropped the missing data and columns.

Secondly, variable/features datatype is key in every data analysis. We had to change wrong datatypes to their approach datatypes to ease analysis.

Thirdly, we had instances of wrong entry in the data. For this, there were two instances, we had a dogtionary entry of '_None_' and also dogs with name '_a_'. We changed both instances to the standard numpy _np.nan_.

Furthermore, we combined the dogtionary columns to become one which makes it easier to group what kind of dog it is according to the dictionary provided.

Finally, we combined the three datasets using the tweet id which is a unique value for each particular tweet.
